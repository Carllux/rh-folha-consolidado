```python
import os
import pdfplumber
from google.colab import drive

# --- 0. CONFIGURA√á√ÉO INICIAL (CAMINHO E DRIVE) ---
# 1. Conecta ao Google Drive
drive.mount('/content/drive')

# 2. Define o caminho exato que voc√™ pediu
caminho_alvo = "/content/drive/MyDrive/Colab Notebooks/_PostalGow/Procedimento RH"

# 3. Tenta entrar na pasta
try:
    os.chdir(caminho_alvo)
    print(f"‚úÖ Diret√≥rio alterado com sucesso para:\n   {os.getcwd()}")
except FileNotFoundError:
    print(f"‚ùå Erro Cr√≠tico: A pasta n√£o foi encontrada.\n   Verifique se o caminho existe no seu Drive:\n   {caminho_alvo}")


# --- 1. FUN√á√ÉO DE AN√ÅLISE (RAIO-X) ---
def raio_x_pdf(caminho_arquivo):
    nome_arquivo = os.path.basename(caminho_arquivo)
    print(f"\n{'='*40}\nüîç ANALISANDO ARQUIVO: {nome_arquivo}\n{'='*40}")

    try:
        with pdfplumber.open(caminho_arquivo) as pdf:
            if len(pdf.pages) > 0:
                pagina = pdf.pages[0]

                # Extra√ß√£o com Layout
                texto_layout = pagina.extract_text(layout=True)

                print("--- VIS√ÉO DO TEXTO (1¬™ P√ÅGINA) ---")
                if texto_layout:
                    print(texto_layout)
                else:
                    print("‚ö†Ô∏è (O texto extra√≠do retornou vazio. Pode ser imagem.)")
                print("-" * 30)
            else:
                print("‚ö†Ô∏è O arquivo PDF parece estar vazio.")

    except Exception as e:
        print(f"‚ùå Erro ao abrir: {e}")


# --- 2. VARREDURA (NO CAMINHO DEFINIDO) ---
print("\nüîÑ Mapeando PDFs na pasta selecionada...")
lista_pdfs = []

# Varre a pasta atual (".") que definimos no passo 0
for root, dirs, files in os.walk("."):
    for file in files:
        if file.lower().endswith(".pdf"):
            # Cria o caminho absoluto para n√£o ter erro de leitura
            path_completo = os.path.abspath(os.path.join(root, file))
            lista_pdfs.append(path_completo)


# --- 3. EXECU√á√ÉO INTELIGENTE (1 EXEMPLO POR PASTA) ---
if lista_pdfs:
    exemplos_por_categoria = {}

    for arquivo in lista_pdfs:
        # Pega o nome da pasta onde o arquivo est√°
        caminho_pasta = os.path.dirname(arquivo)
        nome_pasta = os.path.basename(caminho_pasta)

        # Se ainda n√£o temos um exemplo dessa pasta, salva.
        if nome_pasta not in exemplos_por_categoria:
            exemplos_por_categoria[nome_pasta] = arquivo

    # Relat√≥rio e Execu√ß√£o
    qtd_pastas = len(exemplos_por_categoria)
    print(f"‚úÖ Encontrei {len(lista_pdfs)} PDFs no total.")
    print(f"üîé Analisando 1 exemplo de cada uma das {qtd_pastas} subpastas encontradas...\n")

    for pasta, arquivo in exemplos_por_categoria.items():
        print(f"üìÇ CATEGORIA (PASTA): {pasta.upper()}")
        raio_x_pdf(arquivo)
        print("\n" + "#"*40 + "\n")

else:
    print("‚ùå Nenhum arquivo PDF encontrado dentro de 'Procedimento RH'.")
```

```python
import re
import pandas as pd

dados_extraidos = []

# Regex desenhada para o relat√≥rio "L√≠quido" que vimos no seu log:
# Captura: C√≥digo (n√∫meros) + Espa√ßo + Nome (texto) + CPF + Data + Valor
padrao_liquido = re.compile(r'^\s*(\d+)\s+(.+?)\s+(\d{3}\.\d{3}\.\d{3}-\d{2})\s+(\d{2}/\d{2}/\d{4})\s+([\d\.,]+)')

print("üöÄ Iniciando extra√ß√£o detalhada da pasta 'L√≠quido'...")

# Filtra apenas os arquivos da pasta L√≠quido da sua lista anterior
pdfs_liquido = [arq for arq in lista_pdfs if "L√≠quido" in arq or "Liquido" in arq]

for arquivo in pdfs_liquido:
    try:
        with pdfplumber.open(arquivo) as pdf:
            for pagina in pdf.pages:
                texto = pagina.extract_text()
                if not texto: continue

                # Analisa linha por linha
                for linha in texto.split('\n'):
                    match = padrao_liquido.search(linha)
                    if match:
                        # Se a linha bater com o padr√£o, guarda os dados
                        codigo, nome, cpf, data, valor = match.groups()

                        dados_extraidos.append({
                            "Origem": "L√≠quido",
                            "C√≥digo": codigo,
                            "Funcion√°rio": nome.strip(),
                            "CPF": cpf,
                            "Data Pagto": data,
                            "Valor L√≠quido": valor,
                            "Arquivo": os.path.basename(arquivo)
                        })
    except Exception as e:
        print(f"Erro ao ler {os.path.basename(arquivo)}: {e}")

# Transforma em tabela e mostra
if dados_extraidos:
    df_resultado = pd.DataFrame(dados_extraidos)

    # Converte valor para n√∫mero (tira ponto de milhar e troca v√≠rgula por ponto)
    df_resultado['Valor L√≠quido'] = df_resultado['Valor L√≠quido'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)

    print(f"\n‚úÖ Sucesso! Extra√≠ {len(df_resultado)} registros.")
    display(df_resultado)

    # Salva no Drive
    nome_excel = "Relatorio_Liquidos_Consolidado.xlsx"
    df_resultado.to_excel(nome_excel, index=False)
    print(f"\nüíæ Arquivo salvo no seu Drive como: {nome_excel}")
else:
    print("Nenhum dado capturado. O padr√£o de texto pode ter mudado.")
```

```python
import re
import pandas as pd
import pdfplumber
import os

# 1. Configura√ß√µes Iniciais
dados_extras = []

# --- REGEX (As m√°scaras de leitura) ---
regex_evento = re.compile(r'Lan√ßamentos do Evento:\s*(.*)')
regex_linha = re.compile(r'^\s*(\d+)\s+(.+?)\s+([\d\.,]+)\s+([\d\.,]+)$')

print("üöÄ Iniciando extra√ß√£o da pasta 'EXTRAS'...")

# 2. Filtra apenas os arquivos que parecem ser de Extras
if 'lista_pdfs' not in locals():
    lista_pdfs = []

pdfs_extras = [arq for arq in lista_pdfs if "Extras" in arq or "EXTRAS" in arq]

print(f"üìÇ Encontrei {len(pdfs_extras)} arquivos para processar.\n")

# 3. Processa os PDFs
for arquivo in pdfs_extras:
    nome_arquivo = os.path.basename(arquivo)

    try:
        with pdfplumber.open(arquivo) as pdf:
            evento_atual = "N√£o Identificado"

            for pagina in pdf.pages:
                texto = pagina.extract_text()
                if not texto:
                    continue

                for linha in texto.split('\n'):
                    # Tenta achar o nome do Evento no cabe√ßalho
                    match_evento = regex_evento.search(linha)
                    if match_evento:
                        evento_atual = match_evento.group(1).strip()
                        continue

                    # Tenta ler os dados do funcion√°rio
                    match_dados = regex_linha.search(linha)
                    if match_dados:
                        cod, nome, referencia, valor = match_dados.groups()

                        dados_extras.append({
                            "C√≥digo": int(cod),  # Converte para inteiro
                            "Funcion√°rio": nome.strip(),
                            "Valor (R$)": valor
                        })

    except Exception as e:
        print(f"‚ùå Erro ao ler {nome_arquivo}: {e}")

# 4. Consolida√ß√£o
if dados_extras:
    df_extras = pd.DataFrame(dados_extras)

    # Converte valor para float
    df_extras['Valor (R$)'] = df_extras['Valor (R$)'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)

    print(f"\n‚úÖ Extra√≠ {len(df_extras)} lan√ßamentos de extras.")

    # 5. Agrupa por C√≥digo e soma os valores
    df_agrupado = df_extras.groupby(['C√≥digo', 'Funcion√°rio'], as_index=False)['Valor (R$)'].sum()

    # Ordena pelo valor (decrescente)
    df_agrupado = df_agrupado.sort_values('Valor (R$)', ascending=False)

    print("\nüìä VALORES AGRUPADOS POR C√ìDIGO:")
    print(df_agrupado.to_string(index=False))

    # 6. Salva em CSV/Excel
    df_agrupado.to_excel("valores_agrupados_por_codigo.xlsx", index=False)
    df_agrupado.to_csv("valores_agrupados_por_codigo.csv", index=False)

    print(f"\nüíæ Arquivos salvos:")
    print("   - valores_agrupados_por_codigo.xlsx")
    print("   - valores_agrupados_por_codigo.csv")

else:
    print("‚ö†Ô∏è Nenhum dado capturado.")
```

```python
import os
import pdfplumber
import pandas as pd
import re
from google.colab import data_table

# --- 1. CONFIGURA√á√ÉO E LISTAGEM ---
# Define o caminho da pasta
caminho_pasta = "/content/drive/MyDrive/Colab Notebooks/_PostalGow/Procedimento RH"

try:
    os.chdir(caminho_pasta)
    print(f"‚úÖ Diret√≥rio alterado para: {os.getcwd()}")
except FileNotFoundError:
    print(f"‚ùå Erro: Pasta n√£o encontrada: {caminho_pasta}")

# Lista os arquivos da pasta RECIBO
print("\nüìÇ Mapeando arquivos de Recibo...")
lista_recibos = []
for root, dirs, files in os.walk("."):
    for file in files:
        if "Recibo" in root or "Recibo" in file:
            if file.lower().endswith(".pdf"):
                lista_recibos.append(os.path.abspath(os.path.join(root, file)))

# --- 2. PADR√ïES DE EXTRA√á√ÉO (REGEX) ---
# Captura linha do funcion√°rio: Come√ßa com n√∫mero (c√≥digo), tem texto (nome) e termina com n√∫mero grande (CBO/Empresa)
# Ex: 163 AGATA KETLYN ... 411005
regex_funcionario = re.compile(r'^\s*(\d+)\s+([A-Z\s\.]+?)\s+\d{4,}')

# Captura linha do l√≠quido: Procura por "Total Liquido" e seta (-->)
# Ex: Total Liquido --> 862,00
regex_liquido = re.compile(r'Total Liquido\s*-*>\s*([\d\.,]+)')

# --- 3. EXTRA√á√ÉO ---
dados_recibo = []

print(f"üöÄ Iniciando extra√ß√£o de {len(lista_recibos)} arquivos de Recibo...\n")

for arquivo in lista_recibos:
    try:
        with pdfplumber.open(arquivo) as pdf:
            for pagina in pdf.pages:
                texto = pagina.extract_text()
                if not texto: continue

                # Vari√°veis tempor√°rias para casar Funcion√°rio com seu L√≠quido
                func_temp = None

                for linha in texto.split('\n'):
                    # 1. Procura Funcion√°rio
                    match_func = regex_funcionario.search(linha)
                    if match_func:
                        # Se achou um novo funcion√°rio, guarda os dados
                        cod, nome = match_func.groups()
                        func_temp = {
                            "C√≥digo": cod,
                            "Nome": nome.strip(),
                            "Arquivo": os.path.basename(arquivo)
                        }

                    # 2. Procura Valor L√≠quido (s√≥ se j√° tiver achado um funcion√°rio acima)
                    if func_temp:
                        match_liq = regex_liquido.search(linha)
                        if match_liq:
                            valor = match_liq.group(1)

                            # Adiciona o valor ao registro e salva na lista final
                            func_temp["Valor L√≠quido"] = valor
                            dados_recibo.append(func_temp)

                            # Reseta para buscar o pr√≥ximo recibo na mesma p√°gina (se houver)
                            func_temp = None

    except Exception as e:
        print(f"Erro ao ler {os.path.basename(arquivo)}: {e}")

# --- 4. EXIBI√á√ÉO E SALVAMENTO ---
if dados_recibo:
    df_recibo = pd.DataFrame(dados_recibo)

    # Converte valor para n√∫mero
    df_recibo['Valor L√≠quido'] = df_recibo['Valor L√≠quido'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)

    print(f"\n‚úÖ Sucesso! Extra√≠ {len(df_recibo)} recibos.")

    # Exibe tabela interativa
    data_table.enable_dataframe_formatter()
    display(df_recibo)

    # Salva
    df_recibo.to_excel("Recibos_Consolidados.xlsx", index=False)
    print("\nüíæ Arquivo salvo: Recibos_Consolidados.xlsx")
else:
    print("‚ö†Ô∏è Nenhum dado encontrado. Verifique se os arquivos s√£o PDFs de texto selecion√°vel.")
```

```python
import re
import pandas as pd
import pdfplumber
import os

dados_assistencial = []

# Regex 1: Captura a primeira linha (C√≥digo, Nome, Fun√ß√£o)
# Ex: C√≥digo: 16 Nome : ALESSANDRA... Fun√ß√£o :LIDER...
regex_linha_nome = re.compile(r'C√≥digo:\s*(\d+)\s+Nome\s*:\s*(.+?)\s+Fun√ß√£o\s*:\s*(.*)')

# Regex 2: Captura a linha de baixo (Admiss√£o, Sal√°rio, Valor do desconto)
# Ex: ... Admiss√£o : 04/03/2024 Sal√°rio : 2.487,68 Valor : 10,00
regex_linha_valores = re.compile(r'Admiss√£o\s*:\s*(\d{2}/\d{2}/\d{4})\s*Sal√°rio\s*:\s*([\d\.,]+)\s*Valor\s*:\s*([\d\.,]+)')

print("üöÄ Iniciando extra√ß√£o da 'Rela√ß√£o Assistencial'...")

# Filtra apenas os arquivos da pasta correta
pdfs_assistencial = [arq for arq in lista_pdfs if "Assistencial" in arq]

for arquivo in pdfs_assistencial:
    try:
        with pdfplumber.open(arquivo) as pdf:
            for pagina in pdf.pages:
                texto = pagina.extract_text()
                if not texto: continue

                # Divide o texto em linhas para analisar uma por uma
                linhas = texto.split('\n')

                # Percorre as linhas usando um √≠ndice 'i'
                for i, linha in enumerate(linhas):

                    # Tenta encontrar a Linha 1 (Nome)
                    match_nome = regex_linha_nome.search(linha)

                    if match_nome:
                        # Se achou o nome, vamos tentar pegar os valores na LINHA SEGUINTE (i+1)
                        if i + 1 < len(linhas):
                            linha_baixo = linhas[i+1]
                            match_valores = regex_linha_valores.search(linha_baixo)

                            if match_valores:
                                # Extrai os grupos
                                cod, nome, funcao = match_nome.groups()
                                admissao, salario, valor_desc = match_valores.groups()

                                dados_assistencial.append({
                                    "C√≥digo": cod,
                                    "Funcion√°rio": nome.strip(),
                                    "Fun√ß√£o": funcao.strip(),
                                    "Admiss√£o": admissao,
                                    "Sal√°rio Base": salario,
                                    "Valor Assistencial": valor_desc,
                                    "Arquivo Original": os.path.basename(arquivo)
                                })
    except Exception as e:
        print(f"Erro ao ler {os.path.basename(arquivo)}: {e}")

# Transforma em tabela e mostra
if dados_assistencial:
    df_assistencial = pd.DataFrame(dados_assistencial)

    # Tratamento de n√∫meros (Converter texto '2.487,68' para n√∫mero float)
    cols_numericas = ['Sal√°rio Base', 'Valor Assistencial']
    for col in cols_numericas:
        df_assistencial[col] = df_assistencial[col].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)

    print(f"\n‚úÖ Sucesso! Extra√≠ {len(df_assistencial)} registros da Rela√ß√£o Assistencial.")
    display(df_assistencial)

    # Salva no Drive
    nome_excel = "Relatorio_Assistencial_Consolidado.xlsx"
    df_assistencial.to_excel(nome_excel, index=False)
    print(f"\nüíæ Arquivo salvo no seu Drive como: {nome_excel}")
else:
    print("Nenhum dado capturado. Verifique se os arquivos est√£o na lista.")
```

```python
import re
import pandas as pd
import pdfplumber
import os

dados_folha = []

# --- REGEX (As m√°scaras de leitura) ---
# 1. Gatilho de IN√çCIO: Captura C√≥digo, Nome e Fun√ß√£o
# Ex: C√≥d:16      Nome: ALESSANDRA... Fun√ß√£o:LIDER...
regex_inicio = re.compile(r'C√≥d:\s*(\d+).*?Nome:\s*(.*?)\s+Fun√ß√£o:(.*?)(?:Dep|$)')

# 2. Dados de Contrato: Captura Admiss√£o e Sal√°rio Base
# Ex: Admiss√£o: 04/03/2024 ... Sal√°rio: 2.487,68
regex_contrato = re.compile(r'Admiss√£o:\s*(\d{2}/\d{2}/\d{4}).*?Sal√°rio:\s*([\d\.,]+)')

# 3. Gatilho de FIM: Captura Proventos, Descontos e L√≠quido Final (j√° existentes)
# Ex: Proventos: 3.388,41 Descontos: 475,84 Liquido: 2.912,57
regex_totais = re.compile(r'Proventos:\s*([\d\.,]+).*?Descontos:\s*([\d\.,]+).*?Liquido:\s*([\d\.,]+)')

# 4. Regex para dados da Empresa no cabe√ßalho
regex_razao_social = re.compile(r'(?:Apelido:.*?|\s*)Raz√£o Social:\s*(.*?)(?:\s+CNPJ/CEI:|\s+P√°g:|\n|$)', re.IGNORECASE)
regex_cnpj_cei = re.compile(r'CNPJ/CEI:([\d\./\-]+)', re.IGNORECASE)
regex_periodo = re.compile(r'Per√≠odo de:\s*(\d{2}/\d{2}/\d{4})\s*a\s*(\d{2}/\d{2}/\d{4})', re.IGNORECASE)
regex_endereco = re.compile(r'Endere√ßo:\s*(.*?)(?:\s+N¬∫\s*:|\s+Bairro:|\s+Cidade:|\n|$)', re.IGNORECASE)
regex_numero = re.compile(r'N¬∫\s*:\s*(\d+)', re.IGNORECASE)
regex_complemento = re.compile(r'Compl\.\s*:\s*(.*?)(?:\s+Bairro:|\s+Cidade:|\n|$)', re.IGNORECASE)
regex_bairro = re.compile(r'Bairro\s*:\s*(.*?)(?:\s+Cidade:|\n|$)', re.IGNORECASE)
regex_cidade = re.compile(r'Cidade\s*:\s*(.*?)(?:\s+UF:|\n|$)', re.IGNORECASE)
regex_uf = re.compile(r'UF\s*:\s*([A-Z]{2})', re.IGNORECASE)
regex_cep = re.compile(r'Cep\s*:\s*([\d\-]+)', re.IGNORECASE)

# 5. Novas Regex para campos da 'tabela' (bases e dedu√ß√µes)
regex_base_inss_empresa = re.compile(r'Base INSS Empresa:\s*([\d\.,]+)')
regex_base_inss_funcionario = re.compile(r'Base INSS Funcion√°rio:\s*([\d\.,]+)')
regex_base_inss_func_13o = re.compile(r'Base INSS Func\. 13o\. Sal√°rio:\s*([\d\.,]+)')
regex_base_fgts_13o = re.compile(r'Base F\.G\.T\.S\. 13o\.:\s*([\d\.,]+)')
regex_base_fgts = re.compile(r'Base F\.G\.T\.S\.:\s*([\d\.,]+)')
regex_fgts = re.compile(r'F\.G\.T\.S\.:\s*([\d\.,]+)')
regex_base_irrf = re.compile(r'Base I\.R\.R\.F\.:\s*([\d\.,]+)')
regex_deducoes = re.compile(r'Dedu√ß√µes:\s*([\d\.,]+)')

# 6. Novas Regex para campos do 'cabe√ßalho' (eventos de linha)
# Note que alguns podem aparecer no meio da linha, tentamos ser espec√≠ficos para o valor final
regex_item_salario = re.compile(r'\d+Sal√°rio\s+[\d\.,]+\s+([\d\.,]+)') # Valor do provento Sal√°rio
regex_item_dsr_he = re.compile(r'\d+D\.S\.R\. Sobre Horas Extras\s+([\d\.,]+)')
regex_item_horas_extras_50 = re.compile(r'\d+Horas Extras 50%\s+[\d\.,]+\s+([\d\.,]+)')
regex_item_reembolso_vt = re.compile(r'\d+Reembolso Vale Transporte\s+([\d\.,]+)')
regex_item_inss_salario = re.compile(r'\d+INSS Sobre Sal√°rio\s+[\d\.,]+\s+([\d\.,]+)')
regex_item_irrf_salario = re.compile(r'\d+IRRF Sobre Sal√°rio\s+[\d\.,]+\s+([\d\.,]+)')
regex_item_desc_vt = re.compile(r'\d+Desc\. Vale Transporte\s+[\d\.,]+\s+([\d\.,]+)')
regex_item_contrib_assistencial = re.compile(r'\d+Contribui√ß√£o Assistencial\s+([\d\.,]+)')


print("üöÄ Iniciando extra√ß√£o complexa da 'Folha de Pagamento'...")

# Filtra arquivos da Folha
pdfs_folha = [arq for arq in lista_pdfs if "Folha" in arq]

for arquivo in pdfs_folha:
    try:
        with pdfplumber.open(arquivo) as pdf:
            # --- Extrair informa√ß√µes da empresa uma vez por arquivo ---
            empresa_nome = "N√£o Encontrada"
            empresa_cnpj = "N√£o Encontrado"
            periodo_ini = "N√£o Encontrado"
            periodo_fim = "N√£o Encontrado"
            empresa_endereco = "N√£o Encontrado"
            empresa_numero = "N√£o Encontrado"
            empresa_complemento = "N√£o Encontrado"
            empresa_bairro = "N√£o Encontrado"
            empresa_cidade = "N√£o Encontrado"
            empresa_uf = "N√£o Encontrado"
            empresa_cep = "N√£o Encontrado"

            if pdf.pages:
                first_page_text = pdf.pages[0].extract_text()
                if first_page_text:
                    match_razao_social = regex_razao_social.search(first_page_text)
                    if match_razao_social:
                        empresa_nome = match_razao_social.group(1).strip()

                    match_cnpj_cei = regex_cnpj_cei.search(first_page_text)
                    if match_cnpj_cei:
                        empresa_cnpj = match_cnpj_cei.group(1).strip()

                    match_periodo = regex_periodo.search(first_page_text)
                    if match_periodo:
                        periodo_ini, periodo_fim = match_periodo.groups()

                    match_endereco = regex_endereco.search(first_page_text)
                    if match_endereco:
                        empresa_endereco = match_endereco.group(1).strip()

                    match_numero = regex_numero.search(first_page_text)
                    if match_numero:
                        empresa_numero = match_numero.group(1).strip()

                    match_complemento = regex_complemento.search(first_page_text)
                    if match_complemento:
                        empresa_complemento = match_complemento.group(1).strip()

                    match_bairro = regex_bairro.search(first_page_text)
                    if match_bairro:
                        empresa_bairro = match_bairro.group(1).strip()

                    match_cidade = regex_cidade.search(first_page_text)
                    if match_cidade:
                        empresa_cidade = match_cidade.group(1).strip()

                    match_uf = regex_uf.search(first_page_text)
                    if match_uf:
                        empresa_uf = match_uf.group(1).strip()

                    match_cep = regex_cep.search(first_page_text)
                    if match_cep:
                        empresa_cep = match_cep.group(1).strip()


            for pagina in pdf.pages:
                texto = pagina.extract_text()
                if not texto: continue

                # Vari√°vel tempor√°ria para guardar dados do funcion√°rio atual
                # enquanto lemos as linhas dele
                func_atual = {}

                linhas = texto.split('\n')
                for linha in linhas:

                    # --- PASSO 1: IN√çCIO DO BLOCO (Encontrou "C√≥d:") ---
                    match_inicio = regex_inicio.search(linha)
                    if match_inicio:
                        cod, nome, funcao = match_inicio.groups()
                        func_atual = {
                            "Empresa": empresa_nome,
                            "Empresa CNPJ": empresa_cnpj,
                            "Per√≠odo Inicial": periodo_ini,
                            "Per√≠odo Final": periodo_fim,
                            "Endere√ßo": empresa_endereco,
                            "N√∫mero": empresa_numero,
                            "Complemento": empresa_complemento,
                            "Bairro": empresa_bairro,
                            "Cidade": empresa_cidade,
                            "UF": empresa_uf,
                            "CEP": empresa_cep,
                            "C√≥digo": cod,
                            "Funcion√°rio": nome.strip(),
                            "Fun√ß√£o": funcao.strip(),
                            "Arquivo": os.path.basename(arquivo),
                            # Valores padr√£o caso n√£o encontre nas outras linhas
                            "Admiss√£o": "",
                            "Sal√°rio Base Contratual": "0,00", # Renomeado para evitar conflito
                            "Total Proventos": "0,00",
                            "Total Descontos": "0,00",
                            "L√≠quido a Receber": "0,00",
                            # Novos campos da 'tabela'
                            "Base INSS Empresa": "0,00",
                            "Base INSS Funcion√°rio": "0,00",
                            "Base INSS Func. 13o. Sal√°rio": "0,00",
                            "Base F.G.T.S. 13o.": "0,00",
                            "Base F.G.T.S.": "0,00",
                            "F.G.T.S.": "0,00",
                            "Base I.R.R.F.": "0,00",
                            "Dedu√ß√µes": "0,00",
                            # Novos campos de 'cabe√ßalho' (eventos de linha)
                            "Sal√°rio Provento": "0,00", # Provento '1Sal√°rio'
                            "D.S.R. Sobre Horas Extras": "0,00",
                            "Horas Extras 50%": "0,00",
                            "Reembolso Vale Transporte": "0,00",
                            "INSS Sobre Sal√°rio": "0,00",
                            "IRRF Sobre Sal√°rio": "0,00",
                            "Desc. Vale Transporte": "0,00",
                            "Contribui√ß√£o Assistencial": "0,00"
                        }
                        continue # Vai para a pr√≥xima linha

                    # --- PASSO 2: DADOS CADASTRAIS E OUTROS CAMPOS (S√≥ processa se tiver func_atual aberto) ---
                    if func_atual:
                        match_contrato = regex_contrato.search(linha)
                        if match_contrato:
                            adm, sal_base = match_contrato.groups()
                            func_atual["Admiss√£o"] = adm
                            func_atual["Sal√°rio Base Contratual"] = sal_base

                        # Captura campos da 'tabela'
                        match_base_inss_empresa = regex_base_inss_empresa.search(linha)
                        if match_base_inss_empresa: func_atual["Base INSS Empresa"] = match_base_inss_empresa.group(1)
                        match_base_inss_funcionario = regex_base_inss_funcionario.search(linha)
                        if match_base_inss_funcionario: func_atual["Base INSS Funcion√°rio"] = match_base_inss_funcionario.group(1)
                        match_base_inss_func_13o = regex_base_inss_func_13o.search(linha)
                        if match_base_inss_func_13o: func_atual["Base INSS Func. 13o. Sal√°rio"] = match_base_inss_func_13o.group(1)
                        match_base_fgts_13o = regex_base_fgts_13o.search(linha)
                        if match_base_fgts_13o: func_atual["Base F.G.T.S. 13o."] = match_base_fgts_13o.group(1)
                        match_base_fgts = regex_base_fgts.search(linha)
                        if match_base_fgts: func_atual["Base F.G.T.S."] = match_base_fgts.group(1)
                        match_fgts = regex_fgts.search(linha)
                        if match_fgts: func_atual["F.G.T.S."] = match_fgts.group(1)
                        match_base_irrf = regex_base_irrf.search(linha)
                        if match_base_irrf: func_atual["Base I.R.R.F."] = match_base_irrf.group(1)
                        match_deducoes = regex_deducoes.search(linha)
                        if match_deducoes: func_atual["Dedu√ß√µes"] = match_deducoes.group(1)

                        # Captura campos do 'cabe√ßalho' (eventos de linha)
                        match_item_salario = regex_item_salario.search(linha)
                        if match_item_salario: func_atual["Sal√°rio Provento"] = match_item_salario.group(1)
                        match_item_dsr_he = regex_item_dsr_he.search(linha)
                        if match_item_dsr_he: func_atual["D.S.R. Sobre Horas Extras"] = match_item_dsr_he.group(1)
                        match_item_horas_extras_50 = regex_item_horas_extras_50.search(linha)
                        if match_item_horas_extras_50: func_atual["Horas Extras 50%"] = match_item_horas_extras_50.group(1)
                        match_item_reembolso_vt = regex_item_reembolso_vt.search(linha)
                        if match_item_reembolso_vt: func_atual["Reembolso Vale Transporte"] = match_item_reembolso_vt.group(1)
                        match_item_inss_salario = regex_item_inss_salario.search(linha)
                        if match_item_inss_salario: func_atual["INSS Sobre Sal√°rio"] = match_item_inss_salario.group(1)
                        match_item_irrf_salario = regex_item_irrf_salario.search(linha)
                        if match_item_irrf_salario: func_atual["IRRF Sobre Sal√°rio"] = match_item_irrf_salario.group(1)
                        match_item_desc_vt = regex_item_desc_vt.search(linha)
                        if match_item_desc_vt: func_atual["Desc. Vale Transporte"] = match_item_desc_vt.group(1)
                        match_item_contrib_assistencial = regex_item_contrib_assistencial.search(linha)
                        if match_item_contrib_assistencial: func_atual["Contribui√ß√£o Assistencial"] = match_item_contrib_assistencial.group(1)


                    # --- PASSO 3: FECHAMENTO DO BLOCO (Encontrou "Liquido:") ---
                    if func_atual: # S√≥ busca totais se tiver um funcion√°rio aberto
                        match_totais = regex_totais.search(linha)
                        if match_totais:
                            prov, desc, liq = match_totais.groups()
                            func_atual["Total Proventos"] = prov
                            func_atual["Total Descontos"] = desc
                            func_atual["L√≠quido a Receber"] = liq

                            # ! IMPORTANTE: Adiciona √† lista final e limpa a vari√°vel tempor√°ria
                            dados_folha.append(func_atual)
                            func_atual = {} # Reseta para esperar o pr√≥ximo "C√≥d:"

    except Exception as e:
        print(f"Erro ao ler {os.path.basename(arquivo)}: {e}")

# --- P√ìS-PROCESSAMENTO ---
if dados_folha:
    df_folha = pd.DataFrame(dados_folha)

    # Reordenar as colunas para melhor visualiza√ß√£o
    cols_ordem = [
        "Empresa", "Empresa CNPJ", "Per√≠odo Inicial", "Per√≠odo Final",
        "Endere√ßo", "N√∫mero", "Complemento", "Bairro", "Cidade", "UF", "CEP",
        "C√≥digo", "Funcion√°rio", "Fun√ß√£o", "Admiss√£o", "Sal√°rio Base Contratual",
        "Sal√°rio Provento", "D.S.R. Sobre Horas Extras", "Horas Extras 50%", "Reembolso Vale Transporte",
        "INSS Sobre Sal√°rio", "IRRF Sobre Sal√°rio", "Desc. Vale Transporte", "Contribui√ß√£o Assistencial",
        "Base INSS Empresa", "Base INSS Funcion√°rio", "Base INSS Func. 13o. Sal√°rio",
        "Base F.G.T.S. 13o.", "Base F.G.T.S.", "F.G.T.S.", "Base I.R.R.F.", "Dedu√ß√µes",
        "Total Proventos", "Total Descontos", "L√≠quido a Receber", "Arquivo"
    ]
    df_folha = df_folha[cols_ordem]

    # Converter colunas de valor para n√∫mero
    colunas_valores = [
        "Sal√°rio Base Contratual", "Sal√°rio Provento", "D.S.R. Sobre Horas Extras", "Horas Extras 50%",
        "Reembolso Vale Transporte", "INSS Sobre Sal√°rio", "IRRF Sobre Sal√°rio",
        "Desc. Vale Transporte", "Contribui√ß√£o Assistencial",
        "Base INSS Empresa", "Base INSS Funcion√°rio", "Base INSS Func. 13o. Sal√°rio",
        "Base F.G.T.S. 13o.", "Base F.G.T.S.", "F.G.T.S.", "Base I.R.R.F.", "Dedu√ß√µes",
        "Total Proventos", "Total Descontos", "L√≠quido a Receber"
    ]
    for col in colunas_valores:
        # Garante que a coluna existe antes de tentar processar
        if col in df_folha.columns:
            df_folha[col] = df_folha[col].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)

    print(f"\n‚úÖ Sucesso! Extra√≠ {len(df_folha)} funcion√°rios da Folha.")
    display(df_folha)

    # Salva
    df_folha.to_excel("Folha_Pagamento_Consolidada.xlsx", index=False)
else:
    print("Nenhum dado capturado. Verifique se o termo 'C√≥d:' est√° presente no PDF.")
```

### Mesclando `df_agrupado` (Totais de Extras) com `df_final_merged`

```python
import pandas as pd

# 1. Limpeza de Seguran√ßa: Se j√° existir 'Total Extras' ou 'Valor (R$)' no dataframe, removemos antes
cols_to_drop = [col for col in ['Total Extras', 'Valor (R$)'] if col in df_final_merged.columns]
if cols_to_drop:
    df_final_merged = df_final_merged.drop(columns=cols_to_drop)
    print(f"‚ôªÔ∏è Colunas antigas removidas para evitar duplica√ß√£o: {cols_to_drop}")

# 2. Assegura tipos compat√≠veis
df_final_merged['C√≥digo'] = df_final_merged['C√≥digo'].astype(str).str.strip()
df_agrupado['C√≥digo'] = df_agrupado['C√≥digo'].astype(str).str.strip()

# 3. Merge (Agora seguro)
df_final_merged = pd.merge(
    df_final_merged,
    df_agrupado[['C√≥digo', 'Valor (R$)']],
    on='C√≥digo',
    how='left'
)

# 4. Renomeia
df_final_merged = df_final_merged.rename(columns={'Valor (R$)': 'Total Extras'})

# 5. Preenche vazios (quem n√£o tem extra fica com 0 em vez de NaN)
df_final_merged['Total Extras'] = df_final_merged['Total Extras'].fillna(0.0)

print(f"‚úÖ DataFrames mesclados com sucesso! Colunas atuais: {list(df_final_merged.columns)}")
display(df_final_merged.head())
```
